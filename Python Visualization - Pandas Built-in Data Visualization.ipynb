#### Bar plot

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

```python
df1['A'].plot.hist(bins = 30)
```

#### Scatter plot

```python
df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c = 'C') # set the color based on another column
```

```python
df1.plot.scatter(x='A', y='B', c = 'C', figsize =(10, 5)) 
```

```python
df1.plot.scatter(x='A', y='B', s = df1['C']*50) #add size based on another column
```

#### Boxplot

```python
df2.plot.box() #take every column as a category
```

```python
df = pd.DataFrame(np.random.rand(1000,2), columns = ['a', 'b'])
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25, cmap = 'coolwarm')
```
