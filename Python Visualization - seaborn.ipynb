### Seaborn

- Seaborn is a statistical plotting library
- It has beautiful default styles
- It also is designed to work very well with pandas dataframe objects

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Distribution plots

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
tips.tail(5)
```

```python
sns.distplot(tips['total_bill'])   #plot the distribution
```

```python
sns.distplot(tips['total_bill'], kde = False)   #without kde
```

```python
sns.distplot(tips['total_bill'], kde = False, bins=15) #set the bin values
```

#### Joint plot

- Using Joint plot, you can combine more than one distribution plots.

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips) #use the most
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'hex')  #add kind argument
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'reg')  #add kind  = 'reg', like a linear fitting
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'kde')  
#add kind  = 'kde', shows the density where the dots located the most
#passing into two columns
```

#### Pair plot

- plot as many combination for the numerical columns in the data frame

```python
sns.pairplot(tips)  #a nice way to quickly view the data, passing into the entire dataset
```

```python
sns.pairplot(tips, hue = 'sex')  # hue argument allows us to add categorical variables
                                 #passing the entire dataset
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm' )  # hue argument allows us to add categorical variables
```

#### Matrix plots

- Learn how to create matrix box using seaborn

```python
flights = sns.load_dataset('flights') #loads dataset flights
```

```python
flights.head(5)
```

```python
tips.head(5)
```

In order to make sure the heatmap work properly, the data should be in a matrix form. The index name and columns name matched up. Below is an example -

```python
tc = tips.corr()
print(tc)
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot = True)
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
f = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
print(f)
```

```python
sns.heatmap(f, cmap = 'coolwarm')
```

```python
sns.heatmap(f, cmap = 'magma')
```

```python
sns.heatmap(f, cmap = 'coolwarm', linecolor = 'white', linewidths = 1)
```

```python
sns.clustermap(f, cmap = 'coolwarm') #cluster the data based on similarity
```

```python
sns.clustermap(f, cmap = 'coolwarm', standard_scale = 1) #pass the standard_scale to normalize the scale
```

#### Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(5)
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)  #create the pair plot
```

```python
g = sns.PairGrid(iris)  #PairGrid gives us more control

g.map(plt.scatter)  #create pair scatter plot
```

```python
g = sns.PairGrid(iris)  #PairGrid gives us more control

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips, col = 'time',row = 'smoker')

g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(tips, col = 'time',row = 'smoker')

g.map(sns.scatterplot,'total_bill','tip')  # if pass a plot needs more than one parameter, just pass another variable.
```
